```python
%pylab inline
archivo = genfromtxt("monthrg.dat")
years = array(archivo[:,0] + archivo[:,1]/12)
means = array(archivo[:,3])

x = []
y = []

for i in range (len(years)):
    if means[i] > 0 and years[i] >= 1800 and len(x) < 2048:
        x.append(years[i])
        y.append(means[i])

N = 2**11 
fr = 1/(x[-1]-x[1])
dt = 1./fr / N       
trans = fft.fft(y)
freq = fftfreq(N, dt)

trans[abs(freq) > 0.2] = 0

y2 = ifft (trans)       
figure(figsize= (8,8))
plot(x,y, "cyan")
plot(x, real(y2), color = "red")
#xlim (1800,2000)
ylim (0,250)
show()
```
