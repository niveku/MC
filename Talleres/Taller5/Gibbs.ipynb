```python
%pylab inline

x = linspace(0,2*pi,300)

def Fou(x,num): 
    
    listagrande = []
    for j in range(len(x)):
        lax = x[j]
        suma = 0
        for i in range (2*num):
            if i%2 == 1:
                suma += (sin(i*lax))/i
        listagrande.append(suma)
    return listagrande

figure(figsize = (12,12))
for i in range(3,5):
    y = Fou(x,5*i)
    plot(x,y)
show()    
```
