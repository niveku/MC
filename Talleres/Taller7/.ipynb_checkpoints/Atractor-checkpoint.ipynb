```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

ti = np.linspace(0,600,50000)

a = 0.1
b = 0.1
c = 14

def fun(Y,t):
    # Y[0]=x
    # Y[1]=y
    # Y[2]=z
    return np.array([-Y[1]-Y[2], Y[0]+a*Y[1],b+Y[2]*(Y[0]-c)])
 
sol = odeint(fun,[1.,1.,1.],ti)
 
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
ax.plot(sol[:,0],sol[:,1],sol[:,2],linewidth = 0.5)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.plot (sol[:,0], sol[:,1], linewidth = 0.5)
plt.show()
```

```python
x=[]
y=[]
cs2=np.linspace(1,45,100)
for lac in cs2:
    
    c=lac
    sol = odeint(fun,[1.,1.,1.],ti)
    for j in range(len(sol[10000:,0])):
        if abs(sol[10000+j,0])<=1. and sol[10000+j,1]>=0:
            y.append(sol[10000+j,1])
            x.append(c)
            
plt.scatter(x,y, color = "black", s=0.01)  
plt.title("Rossler System: Bifurcation Diagram.\na=b=0.1")
plt.xlabel("c")
plt.xlim(0,45)
plt.ylim(0,60)
plt.show()
```
