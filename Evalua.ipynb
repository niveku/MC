```python
%pylab inline
from scipy.io import wavfile
ratea = 22050
archivo = genfromtxt("ring.dft", dtype('complex'))

inversa = ifft(archivo)

inversa16 = np.int16(inversa)
print inversa
#wavfile.write('tono.wav',ratea,inversa16)
```

```python
from scipy import interpolate #Importa los módulos necesarios para la interpolación.
xs = linspace (0,len(inversa),len(inversa))
func = interpolate.interp1d(xs, inversa,kind = "linear")
nuevax = linspace (0,len(inversa),2*len(inversa))
ys = func(nuevax)
wavfile.write('tono2.wav',ratea,ys)
figure()
plot(nuevax,ys)
show()
```
