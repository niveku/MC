#Gravedad
El siguiente experimento se llevó a cabo resolviendo estados de masas asociadas a planetas con las ecuaciones diferenciales obtenidad a partir del análisis de la interacción de fuerzas entre ellos.
### Ecuaciones
$F_{12}=-Gm_{1}m_{2}\frac{r1-r2}{||r1-r2||^{3}}=m_{1}a_{1}$

$G=m_{1}=m_{2}=1$

$a_{1}=-\frac{r1-r2}{||r1-r2||^{3}}$

$k=\frac{1}{2}mv^{2} $

$k=\frac{1}{2}v_{x}^{2}+v_{y}^{2}$

$U=\frac{-Gm_{1}m_{2}}{r}$

$U_{12}=\frac{-1}{r_{1}-r_{2}}$

### Nota

Lo anterior sin tener en cuenta que para cada cuerpo actuan más de una fuerza relacionada con las otras masas. Por lo que se implementará la sumatoria de estas dentro del código y sus debidos ajustes.

```python
#Importa librerías necesarias.
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
from matplotlib import animation
%matplotlib inline

#Condiciones iniciales
a= 0.24250109
b = -0.0607718825
c = 0.93240737
d = 0.86473146

#Generación de vectores de posición y velocidad.
r1 = np.array([a,b])
r2 = np.array([-a,-b])
r3 = np.array([0.,0.])
v1 = np.array([c,d])
v2 = np.array([c,d])
v3 = np.array([-2.*c,-2.*d])

Y = np.array([r1,r2,r3,v1,v2,v3]) #Arreglo principal de estados de las masas.

def f(x): # Maneja los vectores y halla los cambios de velocidad producidad por las fuerzas
    R12 = x[0] - x[1] #Gravitatorias de los otros cuerpos.
    R21 = -R12
    R13 = x[0] - x[2]
    R31 = -R13
    R23 = x[1] - x[2]
    R32 = -R23
    A1 = -R12/(norm(R12)**3.) - R13/(norm(R13)**3.) #Basada en la ecución de fuerzas.
    A2 = -R21/(norm(R21)**3.) - R23/(norm(R23)**3.)
    A3 = -R31/(norm(R31)**3.) - R32/(norm(R32)**3.)
    
    return np.array([x[3],x[4],x[5],A1,A2,A3]) #Devuelve la actualización de estados.

#Condiciones de iteraciones y pasaos para Runge-Kutta 
iteraciones=600 
tmax= 0.8
t=0.
Deltat= tmax/iteraciones

def Kutta(): #Maneja las ecuaciones diferenciales y actualiza el estado de las partículas.
    #Basado en el método de Runge-Kutta de órden 4.
    #Permite que las variables sean afectadas fuera del método.
    global t
    global Deltat
    global Y
    #Pasos de Runge-Kutta
    K1 = f(Y)
    K2 = f(Y + Deltat*(1./2.*K1))
    K3 = f(Y + Deltat*(1./2.*K2))
    K4 = f(Y + Deltat*K3)
    Y+= Deltat * (1/6.*K1 + 1/3. * K2 + 1/3.*K3 + 1/6.*K4) #Actualización de los estados.
    t+= Deltat #Actualización del tiempo.
    
history = np.zeros([iteraciones + 1,13]) #Crea un matriz para guardar los estados.
#Inicializa el estado.
history[0] = np.array([t,r1[0],r1[1],r2[0],r2[1],r3[0],r3[1],v1[0],v1[1],v2[0],v2[1],v3[0],v3[1]])

for i in range (int(iteraciones)): #Completa la matríz de historia de estados.
    Kutta()# Da un paso con el método de Runge-Kutta diseñado.
    
    #Extrae los estados de las particulas de la matríz Y actualizada por Kutta().
    x1 = (Y[0])[0]
    y1 = (Y[0])[1]
    vx1 = (Y[3])[0]
    vy1 = (Y[3])[1]
    
    x2 = (Y[1])[0]
    y2 = (Y[1])[1]
    vx2 = (Y[4])[0]
    vy2 = (Y[4])[1]
    
    x3 = (Y[2])[0]
    y3 = (Y[2])[1]
    vx3 = (Y[5])[0]
    vy3 = (Y[5])[1]
    
    history[i+1] = np.array([[t,x1,y1,x2,y2,x3,y3,vx1,vy1,vx2,vy2,vx3,vy3]]) #Guarda el estado.
```

```python
plt.figure(figsize=(12,10))

count = 1
for i in range (3): #Crea las gráficas de las órbitas de las 3 masas.
    
    #Actualiza las X y Y de la masa deseada.
    x = history[:,count]
    count += 1
    y = history[:,count]
    count += 1
    
    plt.subplot(3,1,i+1,aspect = 'equal')
    plt.plot(np.real(x),np.real(y), "--", label= u"Órbita")
    plt.plot(x[0],y[0], "o", label= u"Posición inicial")
    plt.title(u"Órbita M" + str(i+1))
    plt.legend()
    
plt.show()
```

```python
tie = history[:,0]#Lista de los tiempos.
def periodo(): #Busca el periodo de la órbita de los planetas (M1).
    lax = x1 - x1[0] #Parte de 0,0
    lay = x1 - x1[0]
    d = (lax**2+lay**2)**0.5 #Lista de distancias
    elt = 0 #Maneja el mejor valor de t.
    mind = max(d) #Maneja la distancia más cercana.Buscando que sea cercana a 0.
    for i in range (1,len(d)): #Busca la distancia más cercana al punto inicial.
        if d[i] < mind: #Es decir, cuando vuelve a pasar por este punto.
            mind = d[i] #Actualiza los valores
            elt = i
    return elt
periodo = tie[periodo()] #Busca el tiempo del periodo
print "El periodo de los planetas es de: " + ('%.3f' % periodo) + " unidades de tiempo"
```

```python
#Vectores de posición y velocidad iniciales.
v1 = np.array([0.0,0.76])
r1 = np.array([0.384277200514, 0.0])
r2 = np.array([-0.0156823005697, -0.13966430504])
r3 = np.array([-0.352912599375, 0.0])
r4 = np.array([-0.0156823005697, 0.13966430504])
v2 = np.array([-2.01155925929 ,-1.19817066623])
v3 = np.array([0.0, 1.63619158614])
v4 = np.array([2.01155925929,-1.19817066623])
    
Y2 = np.array([r1,r2,r3,r4,v1,v2,v3,v4]) #Lista principal de estados inicial.
#Condiciones de tiempo y pasos.
iteraciones2=100
tmax2= 1.
t2=0.
Deltat2= tmax2/iteraciones2


def update(laa): #Resetea los valores iniciales.
    
    #Permite que las variables se actualizen por fuera del método.
    global r1,r2,r3,r4,v1,v2,v3,v4,Y2,iteraciones2,Deltat2,t2
    
    v1 = np.array([0.0,laa])
    r1 = np.array([0.384277200514, 0.0])
    r2 = np.array([-0.0156823005697, -0.13966430504])
    r3 = np.array([-0.352912599375, 0.0])
    r4 = np.array([-0.0156823005697, 0.13966430504])
    v2 = np.array([-2.01155925929 ,-1.19817066623])
    v3 = np.array([0.0, 1.63619158614])
    v4 = np.array([2.01155925929,-1.19817066623])
    Y2 = np.array([r1,r2,r3,r4,v1,v2,v3,v4])
    iteraciones2=1000
    Deltat2= tmax2/iteraciones2
    t2=0.

def f2(x):# Maneja los vectores y halla los cambios de velocidad producidad por las fuerzas
    R12 = x[0] - x[1] #Producidad por las otras 3 masas.
    R13 = x[0] - x[2]
    R14 = x[0] - x[3]
    R21 = -R12
    R23 = x[1] - x[2]
    R24 = x[1] - x[3]
    R31 = -R13
    R32 = -R23
    R34 = x[2] - x[3]
    R41 = -R14
    R42 = -R24
    R43 = -R34
    
    A1 = -R12/(norm(R12)**3.) - R13/(norm(R13)**3.) - R14/(norm(R14)**3.)
    A2 = -R21/(norm(R21)**3.) - R23/(norm(R23)**3.) - R24/(norm(R24)**3.)
    A3 = -R31/(norm(R31)**3.) - R32/(norm(R32)**3.) - R34/(norm(R34)**3.)
    A4 = -R41/(norm(R41)**3.) - R42/(norm(R42)**3.) - R43/(norm(R43)**3.)
    
    return np.array([x[4],x[5],x[6],x[7],A1,A2,A3,A4])

def Kutta2():#Maneja las ecuaciones diferenciales y actualiza el estado de las partículas.
    #Basado en el método de Runge-Kutta de órden 4.
    #Permite que las variables sean afectadas fuera del método.
    global t2
    global Deltat2
    global Y2
    #Pasos Runge-Kutta
    K1 = f2(Y2)
    K2 = f2(Y2 + Deltat2*(1./2.*K1))
    K3 = f2(Y2 + Deltat2*(1./2.*K2))
    K4 = f2(Y2 + Deltat2*K3)
    Y2+= Deltat2 * (1/6.*K1 + 1/3. * K2 + 1/3.*K3 + 1/6.*K4) #Actualiza los estados.
    t2+= Deltat2 #
```

```python
#Se demora casi 1 minuto.
pasos=60 #maneja la precisión de la búsqueda.
proba=np.linspace(0.70,0.78,pasos) #Crea las posibles a
#Condiciones inciales de mejor a.
mejora = -1
menord = 10000
ide = np.array([0.384277200514, 0.0])#Posición inicial m1
for laa in proba: #Prueba todas las a y su configuración resultante.
    update(laa) #Reinicia el estádo de las partículas
    for i in range (1,int(iteraciones2)): #Aplica los pasos de Runge-Kutta
        Kutta2() #Runge-Kutta 4to órden
        #Halla la distancia con el punto inicial.
        lax = (Y2[0])[0]-ide[0]
        lay = (Y2[0])[1]-ide[1]
        d = lax**2+lay**2
        if d < menord: #Si la distancia es muy corta.
            menord = d
            mejora = laa #Busca la a con mejores resultados para cerrar el ciclo.
print "La a apropiada para el experimento es:" + ('%.5f' % mejora)
```

```python
update(mejora) #Reinicia el estado de las partículas con la a apropiada.

history2 = np.zeros([iteraciones2 + 1,17])#Crea un matriz para guardar los estados.
#Inicializa el estado.
history2[0] = np.array([t2,r1[0],r1[1], r2[0],r2[1], r3[0],r3[1],r4[0],r4[1],
                        v1[0],v1[1], v2[0],v2[1], v3[0],v3[1],v4[0],v4[1]])
for i in range (int(iteraciones2)):#Completa la matríz de historia de estados.
    
    Kutta2()# Da un paso con el método de Runge-Kutta diseñado.
    
    #Extrae los estados de las particulas de la matríz Y actualizada por Kutta().
    x1 = (Y2[0])[0]
    y1 = (Y2[0])[1]
    vx1 = (Y2[4])[0]
    vy1 = (Y2[4])[1]
    
    x2 = (Y2[1])[0]
    y2 = (Y2[1])[1]
    vx2 = (Y2[5])[0]
    vy2 = (Y2[5])[1]
    
    x3 = (Y2[2])[0]
    y3 = (Y2[2])[1]
    vx3 = (Y2[6])[0]
    vy3 = (Y2[6])[1]
    
    x4 = (Y2[3])[0]
    y4 = (Y2[3])[1]
    vx4 = (Y2[7])[0]
    vy4 = (Y2[7])[1]

    #Guarda el estado y su historia.
    history2[i+1] = np.array([[t2,x1,y1,x2,y2,x3,y3,x4,y4,vx1,vy1,vx2,vy2,vx3,vy3,vx4,vy4]])
```

```python
%matplotlib inline
#Historiales de los estados.
tiempo = history2[:,0]
vx1 = history2[:,9]
vy1 = history2[:,10]
vx2 = history2[:,11]
vy2 = history2[:,12]
vx3 = history2[:,13]
vy3 = history2[:,14]
vx4 = history2[:,15]
vy4 = history2[:,16]

# Maneja los estádos de energías.
k = np.zeros(len(vx1)) 
U = np.zeros(len(vx1))

for i in range (len(vx1)):
    #Encuentra la energía cinética de cada instante.
    k1 = 0.5*(vx1[i]**2+vy1[i]**2)
    k2 = 0.5*(vx2[i]**2+vy2[i]**2)
    k3 = 0.5*(vx3[i]**2+vy3[i]**2)
    k4 = 0.5*(vx4[i]**2+vy4[i]**2)
    k[i] = k1+k2+k3+k4
    
    #Encuentra las distancias entre las masas a cada instante.
    RR12= ((x1[i]-x2[i])**2 + (y1[i]-y2[i])**2)**0.5
    RR13= ((x1[i]-x3[i])**2 + (y1[i]-y3[i])**2)**0.5
    RR14= ((x1[i]-x4[i])**2 + (y1[i]-y4[i])**2)**0.5
    RR23= ((x2[i]-x3[i])**2 + (y2[i]-y3[i])**2)**0.5
    RR24= ((x2[i]-x4[i])**2 + (y2[i]-y4[i])**2)**0.5
    RR34= ((x3[i]-x4[i])**2 + (y3[i]-y4[i])**2)**0.5
    
    #Encuentra la energía potencial de cada instante (-1/distancia con otra masa).
    U1 = -1./RR12 -1./RR13 -1./RR14
    U2 = -1./RR12 -1./RR23 -1./RR24
    U3 = -1./RR13-1./RR23 -1./RR34
    U4 = -1./RR14 -1./RR24 -1./RR34
    
    U[i]= 0.5*(U1+U2+U3+U4)

plt.figure()
plt.xlim(0,1)
plt.ylim(-20,30)
plt.xlabel("t")
plt.xlabel("E")
plt.plot(tiempo,k, color = "green", label = u"Energía Cinética")
plt.plot(tiempo,U, color = "blue", label = u"Energía Potencial")
plt.plot(tiempo,U+k, color = "red", label = u"Energía Mecánica")# U+k = la energía Mecánica
plt.legend()
plt.show()
```
