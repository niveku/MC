```python
import turtle as turtle
from Tkinter import *

def dibujo(tortuga, ite, lado):
    
    if ite == 0:       
        
        turtle.fd(lado)
        return
    
    else:
        
        dibujo(turtle, (ite-1) , lado/3)   
        turtle.lt(60)
        dibujo(turtle, (ite-1) , lado/3)   
        turtle.rt(120)
        dibujo(turtle, (ite-1) , lado/3)   
        turtle.lt(60)
        dibujo(turtle, (ite-1) , lado/3)
        

def fractal (tortuga, ite, lar):
    
    tortuga.setup(600,600,0,0)
    tortuga.penup()
    tortuga.setpos(-180,100)
    tortuga.pendown()
        
    for i in range(3):
               
        dibujo(tortuga, ite, lar)
        tortuga.right(120)
```

```python
win = turtle.Screen()
win.tracer(2, 2)
turtle.speed(10)

for i in range(6):
    
    turtle.clear()
    
    lado = 350
    fractal(turtle, i, lado)
    turtle.hideturtle()
    
    
    pantalla = turtle.getscreen()
    pantalla.getcanvas().postscript(file= "fractal" + str(i) + ".png")
    
turtle.exitonclick()
```
