# Koch Snowflafe

```python
import turtle as turtle #importa el directorio de la tortuga.

def dibujo(tortuga, ite, lado): #Método principal de creación de un lado del fractal.
    
    #Inspirado, trabajado y mejorado a partir de discusión en :
    #https://www.daniweb.com/software-development/python/threads/226241/help-koch-snowflake-from-python-2-5-turtle
    
    if ite == 0: # Se encarga de que sea sólo el último nivel el que dibuje el fractal.
        
        turtle.fd(lado) # Momento en el que la tortuga avanza y deja el rastro de largo lado. 
    
    else:
        #Parámetros del dibujo del lado. Donde se manda a un nivel inferior repartiendo el tamaño de los lados entre 3.
        
        dibujo(turtle, (ite-1) , lado/3)   #Este método se llama a si mismo para bajar el nivel de iteración y generar
        turtle.lt(60)                      #los movimientos básicos del fractal de Koch.
        dibujo(turtle, (ite-1) , lado/3)   #Al mismo tiempo reparte el tamaño del lado en (1/3)^n para los lados más 
        turtle.rt(120)                     #pequeños.
        dibujo(turtle, (ite-1) , lado/3)   #Asimismo, se encarga de que la tortuga, la cual lleva la posición y estado
        turtle.lt(60)                      #actual, se transfiera a la siguiente instancia, permitiendo que persista.
        dibujo(turtle, (ite-1) , lado/3)
        

def fractal (tortuga, ite, lar): #Genera los 3 lados del fractal uniendolos de tal forma que cierre el copo de nieve.
    
    tortuga.penup() #Evita que la tortuga dibuje antes de llegar al punto de partida.
    tortuga.setpos(-180,100) #Pone en el punto de partida deseado la tortuga. 
    tortuga.pendown() #Inicia el rastro de la tortuga.
        
    for i in range(3): # Crea los 3 lados (curvas) de Koch
               
        dibujo(tortuga, ite, lar)
        tortuga.right(120)
```

```python
win = turtle.Screen() #Crea 
win.setup(600,600) #Inicializa la venta con dimensiones 600x600 pixeles.

#Manejo de velocidades de graficación.
win.tracer(2, 2)
turtle.speed(0)

for i in range(6): #Crea los 6 distintos fractales.
    
    turtle.clear() #limpia la pantalla y el estado de la tortuga.
    lado = 350
    if i != 0: #Arregla el problema de los tamaños distintos. No tenga ni idea de por qué pero lo hace.
        lado = 350.
    
    fractal(turtle, i, lado) #Crea el fractal con las iteraciones del recorrido.
    turtle.hideturtle() #Esconde la tortuga para que no aparezca en la escena final.
    
    pantalla = turtle.getscreen() #Obtiene el estado actual de la pantalla y la tortuga.
    pantalla.getcanvas().postscript(file= "fractal" + str(i) + ".png") #Guarda la imágen como un archivo PNG.
    
turtle.done() #Indica el final del proceso para la tortuga.
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import misc
import matplotlib.animation as animation

fig = plt.figure(figsize = (15,15))

for i in range (6): #Importa las imágenes de los fractales y crea un panel con éstas.
    
    plt.subplot(3,3,i+1) #Maneja la posición dentro de la gráfica.
    plt.axis('off') #Remueve los ejes.
    png = misc.imread('fractal'+ str(i) +'.png') #Importa la imágen.
    pic = plt.imshow(png, interpolation = "bicubic" ) #Muestra la imágen con interpolación para mejorar su aspecto.

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc
from matplotlib import animation

fig = plt.figure(figsize = (10,10))

def animate(nframe): #Función que importa y plotea las imágenes de los fractales.
    plt.cla() # Limpia la figura antes de plotear la nueva.
    plt.axis('off') #Elimina los ejes.
    #Importa y muestra la imágen.
    png = misc.imread('fractal'+ str(nframe) +'.png')
    plt.imshow(png)
    
    
anim = animation.FuncAnimation(fig, animate, frames=6) #Crea la animación con los 6 fractales.
anim.save('animatedfractal.gif', writer='imagemagick', fps=4, dpi = 200) #Guarda la animación como un GIF animado y maneja su velocidad.
    
```
