```python
lado = 1024 # Define el largo y el ancho del cuadrado de la placa.
centro = lado/2 # Centro de la placa con la rendija.
matriz = zeros ((lado,lado)) # Matriz de la placa inicial.

def cuadrocentro (prop):  # Genera la rendija cuadrada en el centro de la placa con las dimensiones proporcionadas.
    inicio = centro - prop/2
    for i in range (prop): #Recorren sólo el centro de la matriz.
        for j in range(prop):
            matriz[inicio+i][inicio+j] = 1 #Cambia su valor a 1 para que se vea blanco con Imshow()en escala de grises.
        
cuadrocentro(32) # Genera una rendija cuadrada en el centro de la placa con lado 32.
DFT = fft2(matriz) #Aplica la transformada discreta de Fourier a la matriz de la placa.
shift = fftshift (DFT) #Intercambia las frecuencias 0 al centro del espectro.

def sacacuadro(dim): # Método para extraer el cuadrado central de dimensiones 
    cuadrocentral = zeros ((200,200)) # Inicia la matriz del resultado de la interferencia.
    inicio2 = centro - dim/2
    for i in range (dim): #Recorren el centro de la placa.
        for j in range(dim): 
            #Se extraen los valores del centro de la placa aplicando el logaritmo natural al cuadrado del módulo en el punto.
            cuadrocentral [i][j] = log1p (shift[inicio2+i][inicio2+j]**2) #+ (i-100)**2 + (j-100)**2) #Sólo se usa el "z".
    return cuadrocentral #Devuelve la matriz resultada del proceso.
            
interferencia = sacacuadro(200) # Se crea la interferencia con dimensiones 200x200.
#Graficar:
figure(figsize=(14,7)) #Crea la figura para plotear.
subplot(1,2,1) # Inicia la primera división de la figura
imshow(matriz, cmap = "gray") #Plotea la imágen de la placa en escala de grises (B/W).
title(u"Placa con apertura cuadrada")
subplot(1,2,2) # Inicializa la segunda división de la figura.
imshow(real(interferencia)) # Genera la imágen de la interferencia.
title(u"Patrón de interferencia de una apertura cuadrada \n (log resaltado)")
show() # Muestra las gráficas.

#NOTA: No se logró quitar el error -c:21:  
```

```python
lado = 1024 # Define el largo y el ancho del cuadrado de la placa.
centro = lado/2 # Centro de la placa con la rendija.
matriz2 = zeros ((lado,lado)) # Matriz de la placa inicial.
eje1 = 460 # Centro de la primera rendija.
eje2 = 564 # Centro de la segunda rendija.

def rendijas (ax1,ax2,alto,ancho): # Genera 2 rendijas rendijas a partir de sus ejes, altura y ancho.
     for i in range (ax1-ancho/2, ax2+ancho/2): #Recorre sólo lo necesario en la matríz.
        for j in range(centro-alto/2, centro+alto/2):
            # Sólo ingresa los valores dentro de las rendijas.
            if (i>=(ax1-ancho/2) and i<= (ax1+ancho/2)) or (i>=(ax2-ancho/2) and i<= (ax2+ancho/2)): 
                matriz2[j][i] = 1 #Cambia su valor a 1 para que se vea blanco con Imshow()en escala de grises.
        
rendijas(eje1,eje2,200,20) #Crea las rendijas de la placa con los ejes propuestos, alto = 100 y ancho = 20.
DFT2 = fft2(matriz2) #Aplica la transformada discreta de Fourier a la matriz de la placa.
shift2 = fftshift (DFT2) #Intercambia las frecuencias 0 al centro del espectro.

def sacacuadro2(dim): # Genera una interferencia a partir de un sector de la placa.
    cuadrocentral2 = zeros ((dim,dim)) #Inicia la matriz del resultado.
    inicio2 = centro - dim/2
    for i in range (dim): # Recorre sólo lo necesario en el centro de la matriz.
        for j in range(dim):
            cuadrocentral2[i][j] = (shift2[inicio2+i][inicio2+j]**2) #Genera el patrón de interferencia con el valor de intensidad.
    return cuadrocentral2 # Devuelve la matriz con el resultado.
    
difrac = sacacuadro2(128) #Genera la interferencia con dimensiones 128x128
    
#Graficar:
figure(figsize=(14,7)) #Crea la figura para plotear.
subplot(1,2,1) # Inicia la primera división de la figura
imshow(matriz2, cmap = "gray") #Plotea la imágen de la placa en escala de grises (B/W).
title(u"Placa con doble rendija")
subplot(1,2,2) # Inicializa la segunda división de la figura.
imshow(difrac,  interpolation = None) # Genera la imágen de la interferencia.
title(u"Patrón de difracción de una doble rendija")
show() # Muestra las gráficas.
```

```python
lado = 1024 # Define el largo y el ancho del cuadrado de la placa.
centro = lado/2 # Centro de la placa con la rendija.
radio = 95.5 #lado*0.1 # Maneja el radio del orificio en la placa.


#NOTA: El radio descrito en la tarea no genera la misma imágen de esta. La más cercana es u * 0.1 o lado * 0.1


matriz3 = zeros ((lado,lado)) # Genera la matriz de la placa.

def orificio(): # Crea el orificio en el centro de la placa.
    inicio = centro - int(radio) - 1
    for i in range(inicio,inicio+int(2*radio)+1): #Recorre sólo el cuadrante del orificio a crear.
        for j in range (inicio,inicio+int(2*radio)+1):
            if ((i-centro)**2 + (j-centro)**2) <= radio: # Sólo ingresa las posiciones dentro del círculo solicitado.
                matriz3[i][j] = 1 #Cambia su valor a 1 para que se vea blanco con Imshow()en escala de grises.

orificio() # Genera el orificio.
DFT3 = fft2(matriz3) #Aplica la transformada discreta de Fourier a la matriz de la placa.
shift3 = fftshift (DFT3) #Intercambia las frecuencias 0 al centro del espectro.

def interferencia(dim): # Genera la interferencia a partir de un sector de la placa.
    cuadrocentral3 = zeros ((dim,dim)) #Inicia la matriz del resultado.
    inicio3 = centro - dim/2
    for i in range (dim): # Recorre sólo lo necesario en el centro de la matriz.
        for j in range(dim):
            cuadrocentral3[i][j] = (shift3[inicio3+i][inicio3+j]**2) #Genera el patrón de interferencia con el valor de intensidad.
    return cuadrocentral3

interferencia3 = interferencia(lado) # Genera la interferencia con el mismo tamaño de la placa.

linea = 2*pi*0.01/1.*linspace(-512,512,1024)

# Gráfica:
figure(figsize=(12,8)) # Crea la figura.
subplot2grid((3,2),(0,0),colspan=2,rowspan=1) #Genera el primer espacio de la gráfica.
#plotea la gráfica de la comparación de intensidades sin logaritmo.
plot (linea, interferencia3[centro]/interferencia3[centro][centro], linestyle = "-", marker = "*", markersize = 4)
xlim (0,10) # Maneja los límites del eje x para darle el aspecto requerido. 
xlabel("$2 \pi RJ/L$", fontsize= 16)
ylabel("$I/I_{0}$", fontsize= 16)
subplot2grid((3,2),(1,0),rowspan=2) #Genera el segundo espacio de la gráfica para la placa.
imshow(matriz3, cmap = "gray") # Plotea la imágen de la placa en escala de grises.
title(u"Placa con una apertura circular \n")
subplot2grid((3,2),(1,1),rowspan=2) #Genera el Tercer espacio de la gráfica para la interferencia.
imshow(real (log1p (interferencia3))) # Plotea la imágen de interferencia generada a partir de la placa.
title(u"Patrón de interferencia de una apertura circular (Patrón de Airy) \n log resaltado")
tight_layout() # Configura la disposición de las gráficas.
show() # Muestra las gráficas.
```

```python
from scipy import interpolate #Importa los módulos necesarios para la interpolación.
func = interpolate.interp1d(linea, interferencia3[centro]/interferencia3[centro][centro],kind = "cubic") 
#Interpola los datos del eje central y sus intencidades para crear una función.
otrax = linspace(0,4*pi,50) #Permite la gráfica de la función hallada con interpolación.
```

```python
# Límites de la derivación.
a=0
b=4*pi

datos = 500 #Cantidad de datos a tomar en cuenta en la derivación.

unasx = linspace (a,b,datos) # Manejarán las x de las derivadas.
unasy = func(unasx) #Crea una lista con los valores de intensidad en las "x" anteriores.
unasyabajo = roll(unasy, -1) # Mueve los elementos del arreglo a la izquierda. 
h = (b-a)/(datos-1) # Interpreta el espaciamiento de los datos. 
derivada = (unasy-unasyabajo)/h #Halla la derivada por el uso de aproximaciones anteriores.
derivada = derivada[:-1] #Remueve el último elemento ya que, según una gráfica creada, no tiene ningun sentido.
```

```python
figure(figsize=(10,10))
plot(otrax,func(otrax), label = u"Función") #Plotea la cirva de la función hallada con extrapolación.
plot(unasx[1:],derivada, label = "Derivada") #Plotea la curva de la derivada hallada con métodos numéricos.

anterior = [-1000] #Inicia el arreglo que tomará en cuenta el anterior índice que comprometa un min/max.
maxmin = [0] #Indica los índices de las x correspondienes a min/max. Se inicia con 0 porque es máximo indiscutible.
xs = [] #Maneja las abscisas de los min/max.

error = 0.0004 #Tolerancia de la exactitud de la derivada.

for i in range (len (derivada)): #Busca y limpia la lista de índices correspondientes a min/max.
    if derivada[i] < (0. + error) and derivada[i] > ( 0 - error): #Maneja si se acerca lo suficiente a un min/max.
        if anterior[-1] == (i-1):
            anterior.append(i)
        elif anterior[-1] != (i-1): #Se encarga de no repetir índices de un mismo min/max.
            maxmin.append(anterior[int(len(anterior)/2)]) #Agrega el índice del min/max.
            anterior = []
            anterior.append(i)
 
maxmin.remove(-1000) #Remueve el número inicial que evitaba errores.

for j in range (len(maxmin)): #Genera una lista con las abscisas de los min/max a partir de sus índices.
    xs.append(unasx[maxmin[j]])

plot (xs,func(xs), "o",color="red", label = "Max/Min") #Grafíca los min/max.
            
xlim(0,4*pi)
ylim(0,1.2)

legend() #Muestra la leyenda.
show() #Presenta la gráfica resultante.
```

```python
maxi = True #Se encarga de intercambiar máximos y mínimos.
M=1 #Maneja el número de máximo.
m=1 #Maneja el número de mínimo.
for k in range(len(xs)): #Presenta los valores para los puntos mínimos y máximos de intensidad hallados,
    if maxi == True:
        print u"El máximo # " + str(M) + " se encuentra en la abscisa " + ("%.3f" % (xs[k])) + " con un valor de " + ("%.3f" % (func(xs[k])))
        M += 1
        maxi=False
    else:
        print u"El mínimo # " + str(m) + " se encuentra en la abscisa " + ("%.3f" % (xs[k])) + " con un valor de " + ("%.3f" % (func(xs[k])))
        m += 1
        maxi=True 
```
