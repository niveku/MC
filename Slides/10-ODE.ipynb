## Método de Euler y Regla del Punto Medio

$$Y(t_{n+1}) \approx Y(t_n) + f(Y(t_n),t_n) \Delta t$$
$$\textrm{Forward}$$

```python
def f(z):
    return 1.j*z
def one_Euler_step():
    global t
    global Deltat
    global z
    z+= f(z)*Deltat
    t+=Deltat
```

```python
numIters=np.logspace(1,3,20,2)
lastones=[]
plt.figure(figsize=(7,7))
plt.subplot(211)
chosenstep=one_Euler_step
for numIterf in numIters:
    z=1.
    tmax=2*np.pi
    t=0.
    numIter=np.int(numIterf)
    Deltat=tmax/numIter
    history=[[t,z]]
    for __ in range(numIter):
        chosenstep()
        history.append([t,z])
    ahistory=np.array(history)
    lastones.append(history[-1])
    plt.plot(np.real(ahistory[:,1]),np.imag(ahistory[:,1]),label=r"$\Delta t = $"+str(np.round(Deltat,3)))
plt.legend()
plt.title("Euler Method")
plt.axis('equal')
alastones=np.array(lastones)
plt.subplot(212)
plt.plot(tmax/numIters,np.abs(1-alastones[:,1]),"o")
plt.xscale('log')
plt.yscale('log')
plt.show()
np.polyfit(np.log(tmax/numIters),np.log(np.abs(1-alastones[:,1])),1)
```

$$Y(t_{n+1}) \approx Y(t_n) + \frac{1}{2}(K_1+K_2)\Delta t$$
$$ K_1 = f(t_n)$$
$$ K_2 = f( Y + \frac{df}{dY} * f(t_n) \Delta t$$
$$\textrm{Midpoint Rule}$$

```python
def fp(z):
    return 1.j
def one_midpoint_step():
    global t
    global Deltat
    global z
    K1= f(z)
    K2= f(z + f(z)*Deltat) 
    z+= Deltat/2. * (K1+K2)
```

```python
numIters=np.logspace(1,3,20,2)
lastones=[]
plt.figure(figsize=(7,7))
plt.subplot(211)
chosenstep=one_midpoint_step
for numIterf in numIters:
    z=1.
    tmax=2*np.pi
    t=0.
    numIter=np.int(numIterf)
    Deltat=tmax/numIter
    history=[[t,z]]
    for __ in range(numIter):
        chosenstep()
        history.append([t,z])
    ahistory=np.array(history)
    lastones.append(history[-1])
    plt.plot(np.real(ahistory[:,1]),np.imag(ahistory[:,1]),label=r"$\Delta t = $"+str(np.round(Deltat,3)))
plt.legend()
plt.title("Midpoint rule")
plt.axis('equal')
alastones2=np.array(lastones)
plt.subplot(212)
plt.plot(tmax/numIters,np.abs(1-alastones2[:,1]),"o")
plt.xscale('log')
plt.yscale('log')
plt.show()
np.polyfit(np.log(tmax/numIters),np.log(np.abs(1-alastones2[:,1])),1)
```

## Runge-Kutta 3er orden

```python
def one_Kutta3_step():
    global t
    global Deltat
    global z
    K1 = f(z)
    K2 = f(z+Deltat*(1/2.*K1))
    K3 = f(z + Deltat*(-K1+2.*K2))
    z+= Deltat / 6. * (K1+4*K2+K3)
```

```python
numIters=np.logspace(1,3,20,2)
lastones=[]
plt.figure(figsize=(7,7))
plt.subplot(211)
chosenstep=one_Kutta3_step
for numIterf in numIters:
    z=1.
    tmax=2*np.pi
    t=0.
    numIter=np.int(numIterf)
    Deltat=tmax/numIter
    history=[[t,z]]
    for __ in range(numIter):
        chosenstep()
        history.append([t,z])
    ahistory=np.array(history)
    lastones.append(history[-1])
    plt.plot(np.real(ahistory[:,1]),np.imag(ahistory[:,1]),label=r"$\Delta t = $"+str(np.round(Deltat,3)))
plt.legend()
plt.title("Kutta Orden 3")
plt.axis('equal')
alastones3=np.array(lastones)
plt.subplot(212)
plt.plot(tmax/numIters,np.abs(1-alastones3[:,1]),"o")
plt.xscale('log')
plt.yscale('log')
plt.show()
np.polyfit(np.log(tmax/numIters),np.log(np.abs(1-alastones3[:,1])),1)
```

## Runge-Kutta Orden 4

```python
def one_Kutta4_step():
    global t
    global Deltat
    global z
    K1 = f(z)
    K2 = f(z + Deltat*(1/2.*K1))
    K3 = f(z + Deltat*(1./2.*K2))
    K4 = f(z + Deltat*K3)
    z+= Deltat * (1/6.*K1 + 1/3. * K2 + 1/3.*K3 + 1/6.*K4)
```

```python
numIters=np.logspace(1,3,20,2)
lastones=[]
plt.figure(figsize=(7,7))
plt.subplot(211)
chosenstep=one_Kutta4_step
for numIterf in numIters:
    z=1.
    tmax=2*np.pi
    t=0.
    numIter=np.int(numIterf)
    Deltat=tmax/numIter
    history=[[t,z]]
    for __ in range(numIter):
        chosenstep()
        history.append([t,z])
    ahistory=np.array(history)
    lastones.append(history[-1])
    plt.plot(np.real(ahistory[:,1]),np.imag(ahistory[:,1]),label=r"$\Delta t = $"+str(np.round(Deltat,3)))
plt.legend()
plt.title("Kutta Orden 4")
plt.axis('equal')
alastones4=np.array(lastones)
plt.subplot(212)
plt.plot(tmax/numIters,np.abs(1-alastones4[:,1]),"o")
plt.xscale('log')
plt.yscale('log')
plt.show()
np.polyfit(np.log(tmax/numIters),np.log(np.abs(1-alastones4[:,1])),1)
```

```python
def f(Y):
    r3=(np.sqrt(Y[0]**2+Y[1]**2))**3
    return np.array([Y[2],Y[3],-Y[0]/r3,-Y[1]/r3])
```

```python
def one_Kutta4_step():
    global t
    global Deltat
    global Y
    K1 = f(Y)
    K2 = f(Y + Deltat*(1./2.*K1))
    K3 = f(Y + Deltat*(1./2.*K2))
    K4 = f(Y + Deltat*K3)
    Y+= Deltat * (1/6.*K1 + 1/3. * K2 + 1/3.*K3 + 1/6.*K4)
    t+= Deltat
```

```python
def incrementRK4(f,h):
    global Y
    K1 = f(Y)
    K2 = f(Y + h*(1./2.*K1))
    K3 = f(Y + h*(1./2.*K2))
    K4 = f(Y + h*K3)
    return h*(1/6.*(K1 + 2*K2 + 2*K3 + K4))
```

## Multistep Methods

```python
from scipy.special import jv
from scipy.special import jvp
```

```python
# Adams-Bashforth order 4, de esta familia son los métodos utilizados por odeint
# to solve the Bessel differential equation for n = 0
def f(Y,t,n):
    return np.array([Y[1],-1./t*Y[1]-(1.-n**2/t**2)*Y[0]])
```

```python
# f1 is the current slope,
# f2 it the one before
# and f3 the one before that one.
def increment(f1,f2,f3):
    global h
    return  h/12.*(23*f1-16*f2+5*f3)
```

```python
plt.figure(figsize=(10,4))
for n in range(3):
    print "Bessel",n
    numIter=1000
    h=0.01
    Y=np.array([jv(n,h),jvp(n,h)])
    f1 = f(Y,h,n)
    f2 = f1
    f3 = f1
    t=h
    gridSol=np.array([[t,Y[0],Y[1]]])
    x_fine=np.linspace(0,numIter*h,numIter*2)
    
    for __ in range(numIter):
        Y += increment(f1,f2,f3)
        t+=h
        # Actualizar los valores de f1, f2 y f3 para la siguiente iteración
        f1, f2, f3 = f(Y,t,n), f1, f2
        gridSol=np.append(gridSol,[[t,Y[0],Y[1]]],axis=0)
    
    plt.plot(gridSol[:,0],gridSol[:,1],"--",label="sol")
    plt.plot(x_fine,jv(n,x_fine),"k",label="exact ")
    plt.legend()
    #plt.xlim(0,1)
    #plt.ylim(0,2)
plt.show()
```

## Atractor de Lorentz

```python
sigma=10.
rho=28.
beta=8./3.
# Las ecuaciones del sistema de Lorentz son no lineales y acopladas.
def fun(Y,t):
    return np.array([sigma*(Y[1]-Y[0]),Y[0]*(rho-Y[2])-Y[1],Y[0]*Y[1]-beta*Y[2]]) 
```

```python
tmax=50
times=np.linspace(0,tmax,10000)
funsol,others=odeint(fun,[1.,0.,1.],times,full_output=True,printmessg=True)
```
