#9-Integración Numérica

$$ \int_a^b f(x) \,\textrm{d}x \approx \sum_{x_i}w_i f(x_i)$$

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
%matplotlib inline
from scipy import interpolate
```

```python
# Ilustración de la regla trapezoidal y de Simpson
def example_fun(x):
    return x*np.exp(-x)
# Límites de integración
a=0.2
b=1
xvals=np.linspace(a,b,2)
xvals_fine=np.linspace(a-0.01,b+0.01,100)
yvals=example_fun(xvals)
yvals_fine=example_fun(xvals_fine)

fig=plt.figure(figsize=(10,25))

# Regla trapezoidal
ax1=fig.add_subplot(5,1,1)
plt.tick_params(axis='y',labelleft='off',labelbottom='off',labeltop='off')
plt.tick_params(axis='x',labelleft='off',labelbottom='off',labeltop='off')
ax1.plot(xvals_fine,yvals_fine)
ax1.plot(xvals,yvals,"o")
ax1.fill_between(xvals,yvals,alpha=0.4)
ax1.set_title("Trapezoidal Rule")
ax1.text(a,-0.03,r"$x_0$",horizontalalignment="center",fontsize=20)
ax1.text(a-0.06,example_fun(a),r"$f_0$",verticalalignment="center",horizontalalignment="right",fontsize=20)
ax1.text(b,-0.03,r"$x_1$",horizontalalignment="center",fontsize=20)
ax1.text(a-0.06,example_fun(b),r"$f_1$",verticalalignment="center",horizontalalignment="right",fontsize=20)
ax1.plot([a-0.1,b,b],[example_fun(b),example_fun(b),0],"k--",lw=1.5)
ax1.plot([a-0.1,a,a],[example_fun(a),example_fun(a),0],"k--",lw=1.5)
plt.xlim(a-0.05,b+0.05)
xvals=np.linspace(a,b,3)
xvals_fine=np.linspace(a-0.01,b+0.01,100)
xvals_fine2=np.linspace(a,b,100)
yvals=example_fun(xvals)
yvals_fine=example_fun(xvals_fine)

# Regla de Simpson
ax2=fig.add_subplot(5,1,2)
plt.tick_params(axis='y',labelleft='off',labelbottom='off',labeltop='off')
plt.tick_params(axis='x',labelleft='off',labelbottom='off',labeltop='off')
ax2.plot(xvals_fine,yvals_fine)
ax2.plot(xvals,yvals,"o")
interpo=interpolate.lagrange(xvals,yvals)
ax2.fill_between(xvals_fine2,interpo(xvals_fine2),facecolor='red',alpha=0.4)
ax2.set_title("Simpson's rule")
ax2.text(a,-0.03,r"$x_0$",horizontalalignment="center",fontsize=20)
ax2.text(a-0.06,example_fun(a),r"$f_0$",verticalalignment="center",horizontalalignment="right",fontsize=20)
ax2.text((a+b)/2,-0.03,r"$x_1$",horizontalalignment="center",fontsize=20)
ax2.text(a-0.06,example_fun((a+b)/2.),r"$f_1$",verticalalignment="center",horizontalalignment="right",fontsize=20)
ax2.plot([a-0.1,a,a],[example_fun(a),example_fun(a),0],"k--",lw=1.5)
ax2.plot([a-0.1,(a+b)/2.,(a+b)/2.],[example_fun((a+b)/2.),example_fun((a+b)/2.),0],"k--",lw=1.5)
ax2.text(a-0.06,example_fun(b),r"$f_2$",verticalalignment="center",horizontalalignment="right",fontsize=20)
ax2.text(b,-0.03,r"$x_2$",horizontalalignment="center",fontsize=20)
ax2.plot([a-0.1,b,b],[example_fun(b),example_fun(b),0],"k--",lw=1.5)
plt.xlim(a-0.05,b+0.05)

# Mostrar todo junto
plt.show(fig)
```

$$\int_a^b f(x) \textrm{d}x \approx \frac{h}{3} \sum_{j=1}^{N/2} \left[ f(x_{2j-2}) + 4*f(x_{2j-1})+f(x_{2j}) \right] = \frac{h}{3} \left( f(x_0) + 2 \sum_{j=1}^{N/2-1} f(x_{2j}) + 4\sum_{j=1}^{N/2} f(x_{2j-1}) +  f(x_n) \right)$$
$$\textrm{Composite Simpson's Rule}$$

```python
# Estudio del error en función del número de intervalos
# Regla de Simpson
# Requieres an odd number of points so that there be and even number
# of intervals.

def integrand(x):
    return x**2 * np.exp(-x)
a=0.
b=7.
N=10
h = (b-a)/(N-1.)

# Make a plot of the integral to be calculated
x_fine=np.linspace(a,b,100)
x_coarse=np.linspace(a,b,N)
fig=plt.figure(figsize=(10,5))
ax1=fig.add_subplot(1,1,1)
ax1.plot(x_fine,integrand(x_fine))
plt.show()

xvals=np.linspace(a,b,N)
simp = h/3.*(integrand(a) + integrand(b) + 2. * sum(integrand(x_coarse[2:-2:2])) + 4. * sum(integrand(x_coarse[1:-1:2])))
print "h=" + str(h)
print "integral ≈",simp

a=0.
b=7.
teo=(2 + a*(2 + a))/np.exp(a) - (2 + b*(2 + b))/np.exp(b)
print "integral ≈",teo
#teo=(b**3)/3.
inttries=[]
for i in range(2,25,2):
    N = 2.**i
    h = (b-a)/(N)
    x_coarse=np.linspace(a,b,N+1)
    simp = h/3.*(integrand(a) + integrand(b) + 2. * sum(integrand(x_coarse[2:-2:2])) + 4. * sum(integrand(x_coarse[1:-1:2])))
    inttries.append([h,abs(teo-simp)])
ainttries=np.array(inttries)

plt.figure(figsize=(10,5))
plt.plot(ainttries[:,0],ainttries[:,1],"o--")
plt.yscale('log')
plt.xscale('log')
plt.xlabel('$h$')
plt.ylabel('err')
plt.show()
```
