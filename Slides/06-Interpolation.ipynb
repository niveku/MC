## Lagrange

```python
HTML('<iframe src=https://en.wikipedia.org/wiki/Lagrange_polynomial width=600 height=400></iframe>')
```

```python
# Ejemplo de una función que entrega de regreso una función
def lagrangep(absc,j):
    pol=np.poly1d([0,1])
    k=len(absc)
    for m in range(k):
        if m!=j:
            pol*=1./(absc[j]-absc[m])*np.poly1d([1.,-absc[m]])
    return pol   
```

```python
def interlagr(absc,orde):
    poly=np.poly1d([0])
    for i in range(len(absc)):
        poly+=orde[i]*lagrangep(absc,i)
    return poly
```

```python
po=interlagr([0.,1.,2.,4.],[-12.,-12.,-24.,-60.])
```

```python
fig=plt.figure()
ixv=np.linspace(-1,5,100)
ax=fig.add_subplot(111)
for i in range(4):
    ax.plot(ixv,lagrangep([0.,1.,2.,4.],i)(ixv),label="L"+str(i))
for i in range(4):
    ax.plot([0.,1.,2.,4.][i],[1],"o")
    ax.plot([0.,1.,2.,4.][i],[0],"o")
plt.ylim(-2,2)
plt.legend()
plt.title("Polinomios de Lagrange\n",fontsize=20)
plt.show(fig)
```

```python
# Esto ilustra uno de los ejemplos del Survey de Landau
fig=plt.figure()
ixv=np.linspace(-1,4,100)
ax=fig.add_subplot(111)
ax.plot([0.,1.,2.,4.],[-12.,-12.,-24.,-60.],"o",ms=10,label="puntos muestrales")
ax.plot(ixv,po(ixv),label=u"Interpolación")
plt.legend()
plt.xlim(-0.1,4.1)
ax.set_ylim(-60.5,-10)
plt.show(fig)
```

## Newton

```python
HTML('<iframe src=https://en.wikipedia.org/wiki/Newton_polynomial width=600 height=400></iframe>')
```

```python
# Newton basis
def newtonbasis(absc,i):
    poly=np.poly1d([0,1])
    if i!=0:
        for k in range(len(absc)-1):
            poly*=np.poly1d([1,-absc[k]])
    return poly
```

```python
# Divided differences
def div_dif(y,x):
    j=len(x)
    if j==1:
        return y[0]
    else:
        return (div_dif(y[1:],x[1:])-div_dif(y[:-1],x[:-1]))/(x[-1]-x[0])
```

```python
n0=newtonbasis([-1,0,1,2],0)
n1=newtonbasis([-1,0,1,2],1)
n2=newtonbasis([-1,0,1,2],2)
n3=newtonbasis([-1,0,1,2],3)
c0=div_dif([0.],[-1])
c1=div_dif([0.,0],[-1,0],)
c2=div_dif([0.,0,0],[-1,0,1],)
c3=div_dif([0.,0,0,6.],[-1,0,1,2])
po=c0*n0+c1*n1+c2*n2+c3*n3
print np.poly1d(po)
```

```python
# Newton expansion
def newtexp(absc0,ords0):
    absc=[float(i) for i in absc0]
    ords=[float(i) for i in ords0]
    pol=np.poly1d([0])
    for i in range(len(absc)):
        pol+=div_dif(ords[:i+1],absc[:i+1])*(newtonbasis(absc,i))
    return pol
```

```python
po=newtexp([-1,0,1,2],[0,0,0,4])
```

```python
fig=plt.figure()
ixv=np.linspace(-1,2,100)
ax=fig.add_subplot(111)
ax.plot([-1,0,1,2],[0,0,0,4],"o",ms=10,label="puntos muestrales")
ax.plot(ixv,po(ixv),label=u"Interpolación")
plt.legend()
ax.set_ylim(-0.5,1.5)
plt.show(fig)
```

```python
sale = np.genfromtxt(os.path.expanduser("https://raw.githubusercontent.com/ComputoCienciasUniandes/MetodosComputacionalesDatos/master/examples/2015-V/partialsolemio.csv"),delimiter=',')
```

```python
sale_interpol_lin = interpolate.interp1d(sale[:,0],sale[:,1],kind='linear')
sale_interpol_nearest = interpolate.interp1d(sale[:,0],sale[:,1],kind='nearest')
sale_interpol_zero = interpolate.interp1d(sale[:,0],sale[:,1],kind='zero')
sale_interpol_slinear = interpolate.interp1d(sale[:,0],sale[:,1],kind='slinear')
sale_interpol_quadratic = interpolate.interp1d(sale[:,0],sale[:,1],kind='quadratic')
sale_interpol_cubic = interpolate.interp1d(sale[:,0],sale[:,1],kind='cubic')
```

```python
distances = galaxies[:,1] # En kPc
speeds = galaxies[:,2] # En km/s
distances = distances / 1000. # En MPc
# Antes de hacer la gráfica calculemos los parámetros de la regresión lineal
thefit=np.polyfit(distances,speeds,1)
slope=thefit[0]
intercept=thefit[1]
```

```python
#Ejemplo de regresión lineal
plt.figure(figsize=(10,5))
plt.plot(distances,speeds,"o",ms=1)
plt.plot([0,160],[intercept,160*slope+intercept],'c--')
plt.title(u'Ley de Hubble\nEjemplo de regresión lineal',fontsize=20)
plt.xlabel('$\mathrm{distancia}(\mathrm{MPc})$',fontsize=15)
plt.ylabel(r'$\mathrm{rapidez}(\frac{km}{s})$',fontsize=15)
plt.ylim(0)
plt.show()
```

Ejemplo de regresión no lineal.
Datos del laboratorio de Óptica Cuántica: operaciones sobre un qubit
$$\sin ^2(\text{p2}+\text{p3}-x) \sin ^2(2 \text{p1}-\text{p2}-\text{p3}+x)+\cos ^2(\text{p2}+\text{p3}-x)
   \cos ^2(2 \text{p1}+\text{p2}-\text{p3}-x)$$
   p1, p2, y p3 son parámetros

```python
# Ejemplo de regresión no polinomial
# Datos del laboratorio de Óptica Cuántica: operaciones sobre un qubit
quarterdata1=np.genfromtxt(os.path.expanduser("~/Dropbox/Data/blue.csv"),delimiter=',')
quarterdata2=np.genfromtxt(os.path.expanduser("~/Dropbox/Data/red.csv"),delimiter=',')
def func(x,p1,p2,p3):
    return (np.cos(p3 + x - 2*p1 - p2) * np.cos(p3 - x + p2))**2 + (np.sin(p3 - x + p2) * np.sin(p3 - x - 2*p1 + p2))**2
# El último argumento entregado a curve_fit corresponde a los parámetros iniciales
nonlfit=curve_fit(func, quarterdata1[:,0], quarterdata1[:,1],p0=(0,0,0))
plt.figure(figsize=(10,5))
plt.ylabel(r'$P(\theta)$',fontsize=15)
plt.xlabel(r'$\theta$',fontsize=15)
th=np.linspace(0,np.pi,50)
plt.plot(quarterdata1[:,0],quarterdata1[:,1],"or")
plt.plot(quarterdata2[:,0],quarterdata2[:,1],"oc")
#plot(th,func(th,0.4237,0.058,-0.1529),"r")
plt.plot(th,func(th,nonlfit[0][0],nonlfit[0][1],nonlfit[0][2]),"r")
plt.plot(th,1-func(th,0.4237,0.058,-0.1529),'b')
plt.xlim(0,np.pi)
plt.ylim(0,1)
plt.title(u'Ejemplo de regresión no polinómica con datos\n del laboratorio de Óptica Cuántica',fontsize=15)
plt.show()
```

```python
def func(x,p1,p2):
    return p1/(1+p2*np.cos(x))
nonlfit=curve_fit(func,angs,r,p0=(0,0))
a=nonlfit[0][0]
e=nonlfit[0][1]
angs2=np.linspace(0,2*np.pi,100)
radios=func(angs2,a,e)
r=np.sqrt(grigg[:,0]**2+grigg[:,1]**2)
angs=np.arctan2(grigg[:,1],grigg[:,0])

plt.figure(figsize=(10,5))
plt.plot(angs,r,"o")
plt.plot(angs2,radios)
plt.xlabel(r'$\theta$')
plt.ylabel('r/AU')
plt.show()

plt.figure(figsize=(10,5))
plt.plot(grigg[:,0],grigg[:,1],"o",label='puntos orbitales conocidos')
plt.plot(radios*np.cos(angs2),radios*np.sin(angs2),label=u"órbita proyectada")
plt.plot([0.],[0.],"o",label='sol')
plt.xlabel('x/AU')
plt.ylabel('y/AU')
plt.legend()
plt.show()
```
