```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
import matplotlib
import random as random
%matplotlib inline 

alto = 180
ancho = 360
malto = alto/2
mancho = ancho/2
ymin = int(-alto/2)
ymax = int(alto/2)
xmin = int(-ancho/2)
xmax = int(ancho/2)

masa= alto*ancho*0.4
velocidades = [-1,-1,-1,-1,-1,-1,-1,-1,-1,0]
proba = [1,2,-1,-2]
hayrift=False
activo=0
ubirft=[]
mundo = np.zeros([alto,ancho])
#xs = np.linspace(xmin,xmax,ancho) #Coordenadas del cubo.
#ys = np.linspace(ymin,ymax,alto)
#xm,ym = np.meshgrid(xs,ys)  #Matrices de una cara del cubo.

def rift (x):
    global mundo
    global hayrift
    global activo
    for a in range(alto):
        escalera = 0
        if random.random()>0.85:
            escalera = random.choice(velocidades)*random.choice(proba)
        x += escalera
        mundo[a][x]= random.choice(velocidades)
        ubirft.append([x,a,activo])
    activo+=1
    hayrift=True

def actualizarvelocidades():
    global mundo
    for i in range(len(ubirft)):
        mundo [ubirft[i][1]][ubirft[i][0]] = random.choice(velocidades)
        
def initcyclo():
    global mundo
    global ubirft
    ubirft=[]
    mundo = np.zeros([alto,ancho])
    for i in range(alto):
        #lim=random.normalvariate(50, 1)
        for j in range(ancho):
            if (i<50 and j<50):
                mundo[i+malto][j+mancho]=2

    rift(mancho+20)

initcyclo()

# make a color map of fixed colors
colores = colors.ListedColormap(['red','blue','yellow','green'])
limites= np.array([-3,-0.1,1,2,3])
norm = colors.BoundaryNorm(limites, colores.N)
plt.figure(figsize=(8,8))
plt.imshow(mundo, extent=[xmin, xmax, ymin, ymax], cmap = colores, norm= norm)
plt.colorbar()
plt.show()
```

```python
#print len(activo)
def tectonic():
    global mundo
    global hayrift
    if hayrift:
        for i in range(alto):
            r = np.zeros(activo)
            actual=0
            for j in range(ancho):
                if mundo[i][j] < 0:
                    r[actual] = j
                    #print actual,j,r, mundo[i][j+1]
                    actual += 1
            nuevafila=np.zeros(ancho)     
            for j in range(ancho):
                dif=0
                if mundo[i][j] < 0:
                    nuevafila[j] = mundo[i][j] 
                elif mundo[i][j] > 0:
                    for rift in r:
                        if abs(rift-j)>600:
                            hayrift=False
                        if rift-j>0:
                            dif += mundo[i][rift]
                        elif rift-j<0:
                            dif-= mundo[i][rift]
                        nuevafila[(j+dif)%ancho]=mundo[i][j]
                elif mundo[i][j] == 0:
                    if (mundo[i][(j+1)%ancho] == 2 or mundo[i][(j-1)%ancho] == 2):
                        nuevafila[(j+dif)%ancho] = 1
                    elif(mundo[(i+1)%alto][j] == 2 or mundo[(i-1)%alto][j] == 2):
                        nuevafila[(j+dif)%ancho] = 1 
            mundo[i]=nuevafila
            
    elif hayrift == False:
        for i in range(alto):
            for j in range(ancho):
               if mundo[i][j] == 0 and mundo[i][(j+1)%ancho] == 1:
                    mundo[i][j] == -1
            
    actualizarvelocidades()
tectonic()       
plt.figure(figsize=(8,8))
plt.imshow(mundo, extent=[xmin, xmax, ymin, ymax], cmap = colores, norm= norm)
#plt.colorbar()
plt.show()    
```

```python
%matplotlib tk
import matplotlib.animation as animation

def anim(t): #Función para actualizar la animación.
    global mundo #Permite alterar el estado del mundo.
    tectonic() #Crea un instante nuevo de las placas.
    plt.title(str(t*0.1)+"Ma")
    globo.set_array(mundo) #Actualización de la gráfica.
    return [globo]

fig = plt.figure() #Figura.
ax = plt.axes()
initcyclo()
globo = plt.imshow(mundo, extent=[xmin, xmax, ymin, ymax], cmap = colores, norm= norm)

anim = animation.FuncAnimation(fig, anim, frames=100 , interval=10, blit=True) #Animación.
anim.save('placas.gif',writer='imagemagick', fps = 5)
#plt.show()
```
