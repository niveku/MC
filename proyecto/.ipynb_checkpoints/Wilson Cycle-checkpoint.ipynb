#Wilson Cycle

### Introducción

#### Tectónica de placas.
La [Tectónica de placas](https://es.wikipedia.org/wiki/Tect%C3%B3nica_de_placas) es la teoría base de la geología moderna, la cual se basa en el movimiento e interacción de [placas tectónicas](https://es.wikipedia.org/wiki/Placa_tect%C3%B3nica), lo cual es responsable de bastantes fenómenos en la tierra como terremotos, volcánes y orogénias. 
![](https://upload.wikimedia.org/wikipedia/commons/thumb/5/56/Placas_tectonicas_mayores.svg/300px-Placas_tectonicas_mayores.svg.png)


Los [límites de placas](http://recursos.cnice.mec.es/biosfera/alumno/4ESO/MedioNatural1I/contenido3.htm) puedes ser de 3 tipos:

1. **Divergente:** Que son los lugares de rompimiento de continentes(["rifting"](https://es.wikipedia.org/wiki/Rift)) o las [dorsáles centro-oceánicas](https://es.wikipedia.org/wiki/Dorsal_centro-oce%C3%A1nica) (MOR por sus siglas en Inglés). En ambos casos se presenta ascenso del manto de la tierra lo que genera nuevo material en la superficie.
![](http://1.bp.blogspot.com/-bgh-3oREUCA/Tk1lB8UnAFI/AAAAAAAAACI/foHTbaIqysA/s1600/BORDE+DIVERGENTE.png)
2. **Convergente:** Donde se encuentran dos placas distintas. Se puede dar el caso de que una placa oceánica comienze a irse debajo de una placa continental produciendo una [subducción](https://es.wikipedia.org/wiki/Subducci%C3%B3n) o que se encuentre dos masas terrestres provocando una colisión.
![](http://www.ediciona.com/portafolio/image/8/4/3/8/subduccion_y_formacion_islas_8348.jpg)
3. **Transformante:** Donde se encuentran dos placas tectónicas con velocidades paralelas. Existe gran cantidad de estas fallas en los MOR produciendo un "escalonamiento" típico de este sistema. 
![](http://roble.pntic.mec.es/afep0032/imagenes/_pasivo.jpg)

#### Cyclo de Wilson

El [ciclo de Wilson](http://www.le.ac.uk/gl/art/gl209/lecture3/lecture3.html) está basado a partir de los pilares de la tectónica de placas, cambia radicalmente la configuración de los terrenos involucrados debido a la acción de ruptura, subducción, orogénia y colisión. Por esta razón es que se presentan cambios de elevación en los terrenos, usualmente asociados con vulcanismo.

![](http://blog.sciencenet.cn/data/attachment/album/201507/01/195827t09y0yvk9s07kkmk.jpg)

#### Objetivo

Generar una matriz que represente una porción de un planeta donde haya un continente cuadrado (para comparación) de relieve uniforme a partir del cual se generara un ciclo de Wilson en el que se verá comprometido el relieve de dicha porción de tierra. Esto con el fin de ejemplificar de forma simple como procesos que transcurren durante miles de años en la tierra logran cambiar su apariencia. 

**Cambios de relieve**

Los cambios de relieve que se tratarán de forma simplificada son la expansión de las playas debido a la erosión de los terrenos continentales, la orogenía asociada a vulcanismo producida por el ascenso de material caliente en una subducción.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
import matplotlib
import random as random
%matplotlib inline 

alto = 180
ancho = 360
malto = alto/2
mancho = ancho/2
ymin = int(-alto/2)
ymax = int(alto/2)
xmin = int(-ancho/2)
xmax = int(ancho/2)

masa= alto*ancho*0.4
velocidades = [-1,-1,-1,-1,-1,-1,-1,-1,-1,0]
proba = [1,2,-1,-2]
hayrift=False

activo=0
ubirft=[]
mundo = np.zeros([alto,ancho])
#xs = np.linspace(xmin,xmax,ancho) #Coordenadas del cubo.
#ys = np.linspace(ymin,ymax,alto)
#xm,ym = np.meshgrid(xs,ys)  #Matrices de una cara del cubo.

def rift (x):
    global mundo
    global hayrift
    global activo
    for a in range(alto):
        escalera = 0
        if random.random()>0.85:
            escalera = random.choice(velocidades)*random.choice(proba)
        x += escalera
        mundo[a][x]= random.choice(velocidades)
        ubirft.append([x,a,activo])
    activo+=1
    hayrift=True

def actualizarvelocidades():
    global mundo
    for i in range(len(ubirft)):
        mundo [ubirft[i][1]][ubirft[i][0]] = random.choice(velocidades)

def estadorift(a):
    global mundo
    for i in range(len(ubirft)):
        mundo [ubirft[i][1]][ubirft[i][0]] = a
        
def initcyclo():
    global mundo
    global ubirft
    global hayrift
    ubirft=[]
    mundo = np.zeros([alto,ancho])
    for i in range(alto):
        #lim=random.normalvariate(50, 1)
        for j in range(ancho):
            if (i<50 and j<50):
                mundo[i+malto][j+mancho]=2

    rift(mancho+20)

initcyclo()

# make a color map of fixed colors
colores = colors.ListedColormap(['red','blue','yellow','green'])
limites= np.array([-3,-0.1,1,2,3])
norm = colors.BoundaryNorm(limites, colores.N)
plt.figure(figsize=(8,8))
plt.imshow(mundo, extent=[xmin, xmax, ymin, ymax], cmap = colores, norm= norm)
#plt.colorbar()
plt.show()
```

```python
initcyclo()

def mariz():
    global mundo
    lista = np.ones(alto, dtype=int)
    for i in range(alto):
        for j in range(ancho):
            if j > ubirft[i][0] and mundo[i][j] == 0 and mundo[i][(j+1)%ancho] > 0:
                mundo[i][j] = -1
                lista[i]=j
    return lista

sub = False

def tectonic():
    global mundo
    global hayrift
    global subduce
    global sub
    if hayrift:
        for i in range(alto):
            nuevafila=np.zeros(ancho)     
            for j in range(ancho):
                dif=0
                if mundo[i][j] == 0:
                    if (mundo[i][(j+1)%ancho] == 2 or mundo[i][(j-1)%ancho] == 2):
                        nuevafila[(j+dif)%ancho] = 1
                    elif(mundo[(i+1)%alto][j] == 2 or mundo[(i-1)%alto][j] == 2):
                        nuevafila[(j+dif)%ancho] = 1
                elif mundo[i][j] < 0:
                    nuevafila[j] = mundo[i][j] 
                elif mundo[i][j] > 0:
                    if ubirft[i][0]-j>0:
                        dif += mundo[i][ubirft[i][0]]
                    elif ubirft[i][0]-j<0:
                        dif-= mundo[i][ubirft[i][0]]
                    nuevafila[(j+dif)%ancho]=mundo[i][j] 
            mundo[i]=nuevafila
            actualizarvelocidades()
            
    elif sub == True:
        for i in range(alto):
            fila = np.zeros(ancho)
            print "HAY"
            for j in range(0,subduce[i]-1):                
                if mundo[i][j] > 0:
                    fila[i][j+5]=mundo[i][j]
            for j in range(subduce[i],ancho):
                fila[i][j+5]=mundo[i][j]
            mundo[i]=fila
        
    elif hayrift == False and sub == False:
        estadorift(0)
        print "Check1"
        subduce = np.array(mariz())
        print "Check2"
        sub = True
        print "Check3"
        print subduce
    
tectonic()       
plt.figure(figsize=(8,8))
plt.imshow(mundo, extent=[xmin, xmax, ymin, ymax], cmap = colores, norm= norm)
#plt.colorbar()
plt.show()    
```

```python
%matplotlib tk
import matplotlib.animation as animation

def anim(t): #Función para actualizar la animación.
    global mundo #Permite alterar el estado del mundo.
    global hayrift
    plt.title(str(t*0.1)+"Ma")
    globo.set_array(mundo) #Actualización de la gráfica.
    if t*0.1 == .9:
        hayrift=False
    tectonic() #Crea un instante nuevo de las placas.
    return [globo]

fig = plt.figure() #Figura.
ax = plt.axes()
plt.xlabel("Km")
plt.ylabel("Km")
initcyclo()
globo = plt.imshow(mundo, extent=[xmin, xmax, ymin, ymax], cmap = colores, norm= norm)

anim = animation.FuncAnimation(fig, anim, frames=200 , interval=0, blit=True) #Animación.
#anim.save('placas.gif',writer='imagemagick', fps = 6, dpi=20)
#plt.show()
```

###Conclusiones.
Se puede observar como a partir de un ridge/rift intracontinental se puede abrir un oceano, el cuál puede ser eliminado gracias al cierre de la abertura por obra de la subducción, en este caso de un sólo lado del continente derecho.
Este ciclo es conocido como el ciclo de wilson y es un de los eventos más interesantes en geología ya que permite ciertas configuraciones de rocas en orogénias caracterizadas por la presencia de material.

Por lo anterior, se observó que el ciclo de wilson, programado para iniciar su cierre a **\__ Ma** de producirse la ruptura cambio notoriamente el relieve del continente cuadrado inicial.

Finalmente, es notoria la gran cantidad de simplificación realizada para que fuera viable una programación no muy exigente computacionalmente. Sin embargo, el código puede ser ampliamente mejorado y adaptado a otras condiciones.
