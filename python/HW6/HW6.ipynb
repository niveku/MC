```python
import sympy as sympy #Importa la librería de Sympy
sympy.init_printing(use_unicode=True)# Corrije la presentación del output.

def lagrange(n):
    if (n >= 1):#Evita errores con la procesos.
        f = sympy.symarray('f',n) # Lista de f a usar (con subíndices).
        t = sympy.symarray('t',n) # Lista de t a usar (con subíndices).
        symt = sympy.symbols('t') # Símbolo t.
        polinomio = 0 # Inicia la respuesta que se devolverá.
        for i in range(n): #Recorre n veces para completar el polinomio.
            base = 1 #Inicia el múltiplo.
            for j in range(n): #Recorre n veces buscando diferentes índices.
                if(i!=j): 
                    base *=(symt - t[j])/(t[i] - t[j])
            polinomio += f[i]*base #Agrega el término deseado.
            
    return sympy.collect(polinomio, f) #Asocia la respuesta a las f. 

def adams(n):
    if (n >= 1):#Evita errores con la procesos.
        h = sympy.symbols('h') # Símbolo h para el paso.
        nt = sympy.symarray('t',n) # Lista de t a usar (con subíndices).
        symt = sympy.symbols('t') # Símbolo t.
        polinomio = sympy.simplify(lagrange(n)) # Polinómio de Lagrange correspondiente usado para la sustitución.
        polinomio = sympy.simplify(sympy.integrate(polinomio, (symt, nt[0], nt[0]+h) ) )
        for i in range(n): #Recorre n veces para arreglar el polinomio.
           # Se realiza la modificación necesaria en el polinomio para su operación.
            polinomio = polinomio.subs(nt[i],nt[0]-h*(i)) 
        # Se integra (en el rango t[0] hasta t[0]+h) y simplifica el polinómio respuesta.
        return sympy.simplify(polinomio) 
```

```python
%pylab inline
#Constantes

d=400. #Densidad volumétrica de la masa (kg/m^3).
l=270 #Longitud de un lado del cubo de la masa (km).
cubo=3030 #Longitud del cubo grande contenedor (km).
divisiones = 101. #Para que la longitud de cada segmento sea 30m. 
centro = 51#Índice central del cubo.
#Índice de los extremos del cubo pequeño.
a = centro-4 
b = centro+4
#Límites del cubo grande.
xmin= -cubo/2. 
xmax = cubo/2. 
xs = linspace(xmin,xmax,divisiones) #Coordenadas del cubo.
xm,ym,zm = np.meshgrid(xs,xs,xs)  #Matrices del cubo.
campo = zeros((divisiones,divisiones,divisiones)) #Inicialización del campo gravitatório.
```

```python
campo[a:b,a:b,a:b]= 1.
iteraciones = 1000 #Numero de iteraciones para estimar el campo.

for __ in range(iteraciones): #Completa el campo.
    #Actualiza el campo en los puntos con el promedio con los valores que lo rodean.
    campo = 1/6.*(np.roll(campo,1,axis=0)+np.roll(campo,-1,axis=0)+np.roll(campo,1,axis=1)+
                np.roll(campo,-1,axis=1)+np.roll(campo,1,axis=2)+np.roll(campo,-1,axis=2))
    #Condiciones de borde.
    campo[0] = 0.
    campo[-1] = 0.
    campo[:,0] = 0.
    campo[:,-1] = 0.
    campo[:,:,0] = 0.
    campo[:,:,-1] = 0.
```

```python
figure(figsize = (9.5,8))
pcolor(1e-4*campo[centro,:,:],cmap='gist_heat_r') #Plotea el campo en unidades (cm^2/s^2)
#x fija en el centro y valores del plano yz = [centro,:,:]
xlabel(r"$y \ (km)$",fontsize=15)
ylabel(r"$z \ (km)$",fontsize=16)
title(u"$\Phi\\ cm^{2}s^{-2}$",fontsize=18)
colorbar()#Presenta la barra lateral con las convenciones de colores.
#Genera contornos (de nivel) que indican cierta magnitud de campo. 
contornos = contour(1e4*campo[centro,:,:], linestyles = "--",colors = "black")
#Escribe los valores del campo en los contornos.
clabel(contornos, inline=1, fontsize=10)
show()
```

```python
zf = roll(1e5*campo[centro,centro],-1) #Halla los valores de campo siguientes en el eje z.
derivada = zf-1e5*campo[centro,centro] #Encuentra la diferencia de los valores en un paso.
figure(figsize=(10,7))
title("Campo Gravitacional")
xlabel("$z\ (km)$",fontsize=15)
ylabel("$g \ (\ \mu m \ /s^{2})$", fontsize=15)
#Borra los datos no deseados y plotea los resultantes.
scatter(xs[:-1],derivada[:-1],facecolor='none') 
xlim(xmin,xmax)
show()
```

```python
from scipy import interpolate #Importa los módulos necesarios para la interpolación.
#Interpolación cúbica para hallar valores de campo en función de la posición.
acel = interpolate.interp1d(xs[:-1],derivada[:-1]/1e6,kind = 3) 
#Interpola los datos del eje central y sus intencidades para crear una función.
otrax = linspace(xs[0],xs[-2],100) #Permite la gráfica de la función hallada con interpolación.
figure(figsize=(10,7))
title(u"Interpolación")
xlabel("$z\ (km)$",fontsize=15)
ylabel("$g \ ( m \ /s^{2})$", fontsize=15)
xlim(xmin,xmax)
plot(otrax,acel(otrax))
show()
```

```python
y = [500.]#Lista de las posiciones en los instantes (m).
v = [0.]#Lista de las velocidades en los instantes (m/s).
a = [acel(y[-1])] #Lista de las aceleraciones en los instantes (m/s^2).
lim=(l/2.) # Indíca el límite donde se encuentran los cuerpos.

def buscat(): #Busca el tiempo en el cual se chocan los objetos.
    i = 0.
    #Este método se basa en la división de instantes de la ecuación de movimiento
    #de caida libre, ya que la otra masa es muy pequeña comparada con el cubo central.
    while y[-1] > lim:
        #Agrega los nuevos valores de posición, velocidad y aceleración.
        y.append(y[-1] + v[-1] + 0.5*a[-1])
        v.append(v[-1]+acel(y[-2]))
        a.append(acel(y[-2]))
        # Se maneja de tal forma que no se necesita el tiempo ya que siempre es 1 seg.
        i += 1. #Avanza un instante (seg).
    return i #Devuelve el tiempo de choque.

lat = buscat() #Encuentra el tiempo de choque.
#Maneja el tiempo de forma que presente horas, minutos y segundos.
hora = int(lat/3600.)
minuto = int(lat/60 - hora*60) #Menos los tomados dentro de horas.
segundos = lat-hora*3600-minuto*60 #Menos los tomados dentro de horas y minutos.

print (u"\n La partícula colisionará después de " + str(hora) + " Horas "+str(minuto) +
" minutos y " + str (int(segundos)) + " segundos.") #Imprime la respuesta.
        
figure(figsize=(10,12))

#Plote la posición, velocidad y aceleración cpntra el tiempo.

subplot(3,1,1)
title(u"Posición")
ylabel("Distancia $(m)$",fontsize=12)
xlim(0,lat+1)
plot(arange(len(y)),y)
plt.axhline(y=lim, xmin=0.0, xmax=1, linestyle= "--",linewidth=1, color = 'k')

subplot(3,1,2)
title(u"Rapidez")
ylabel("Rapidez $(m/s)$",fontsize=12)
xlim(0,lat+1)
plot(arange(len(v)),[abs(vs) for vs in v])

subplot(3,1,3)
plot(arange(len(v)),a)
xlim(0,lat+1)
title(u"Aceleración")
ylabel(u"Aceleración $(m/s^{2}$)",fontsize=12)
xlabel("Tiempo (s)",fontsize=13)
show()
```
